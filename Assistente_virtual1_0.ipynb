```python
!pip install SpeechRecognition gTTS wikipedia

import speech_recognition as sr
import wikipedia
from gtts import gTTS
import os
from IPython.display import Audio, display, HTML
from google.colab import files
```

```python
def falar(texto, filename="resposta.mp3"):
    """Converte texto em fala e reproduz no Colab"""
    tts = gTTS(text=texto, lang="pt")
    tts.save(filename)
    display(Audio(filename, autoplay=True))
```

```python
def ouvir():
    """Faz upload de um arquivo .wav e converte fala em texto"""
    print("üìÇ Fa√ßa upload de um arquivo de √°udio (.wav)")
    uploaded = files.upload()

    for fn in uploaded.keys():
        recognizer = sr.Recognizer()
        with sr.AudioFile(fn) as source:
            audio = recognizer.record(source)
        try:
            texto = recognizer.recognize_google(audio, language="pt-BR")
            print(f"üó£Ô∏è Voc√™ disse: {texto}")
            return texto.lower()
        except sr.UnknownValueError:
            print("‚ùå N√£o entendi o √°udio.")
            return ""
        except sr.RequestError:
            print("‚ö†Ô∏è Erro na API do Google.")
            return ""
```

```python
def executar_comando(comando):
    if "wikipedia" in comando:
        falar("O que voc√™ deseja pesquisar na Wikip√©dia?")
        termo = ouvir()
        if termo:
            try:
                resumo = wikipedia.summary(termo, sentences=2, auto_suggest=False)
                print("üìñ Resumo da Wikip√©dia:", resumo)
                falar(resumo)
                display(HTML(f'<a href="https://pt.wikipedia.org/wiki/{termo}" target="_blank">üîó Abrir Wikipedia sobre {termo}</a>'))
            except:
                falar("N√£o encontrei nada sobre isso.")

    elif "youtube" in comando:
        falar("Abrindo o YouTube, clique no link abaixo.")
        display(HTML('<a href="https://www.youtube.com" target="_blank">üîó Clique aqui para abrir o YouTube</a>'))

    elif "farm√°cia" in comando:
        falar("Abrindo a farm√°cia mais pr√≥xima no Google Maps, clique no link abaixo.")
        display(HTML('<a href="https://www.google.com/maps/search/farm√°cia+perto+de+mim/" target="_blank">üîó Clique aqui para abrir farm√°cias pr√≥ximas</a>'))

    elif "sair" in comando:
        falar("At√© logo!")
        return False

    else:
        falar("N√£o entendi o comando.")

    return True
```

```python
falar("Ol√°, eu sou sua assistente virtual. Como posso ajudar?")

ativo = True
while ativo:
    comando = ouvir()
    if comando:
        ativo = executar_comando(comando)
```
