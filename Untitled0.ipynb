```python
!pip install SpeechRecognition gTTS wikipedia

import speech_recognition as sr
import webbrowser
import wikipedia
from gtts import gTTS
import os
from IPython.display import Audio, display
```

```python
def falar(texto, filename="resposta.mp3"):
    tts = gTTS(text=texto, lang="pt")
    tts.save(filename)
    display(Audio(filename, autoplay=True))
```

```python
from google.colab import files

def ouvir():
    print("üìÇ Fa√ßa upload de um arquivo de √°udio (.wav)")
    uploaded = files.upload()

    for fn in uploaded.keys():
        recognizer = sr.Recognizer()
        with sr.AudioFile(fn) as source:
            audio = recognizer.record(source)
        try:
            texto = recognizer.recognize_google(audio, language="pt-BR")
            print(f"Voc√™ disse: {texto}")
            return texto.lower()
        except sr.UnknownValueError:
            print("N√£o entendi o √°udio.")
            return ""
        except sr.RequestError:
            print("Erro na API do Google.")
            return ""
```

```python
def executar_comando(comando):
    if "wikipedia" in comando:
        falar("O que voc√™ deseja pesquisar na Wikip√©dia?")
        termo = ouvir()
        if termo:
            try:
                resumo = wikipedia.summary(termo, sentences=2, auto_suggest=False)
                print(resumo)
                falar(resumo)
            except:
                falar("N√£o encontrei nada sobre isso.")

    elif "youtube" in comando:
        falar("Abrindo o YouTube...")
        webbrowser.open("https://www.youtube.com")

    elif "farm√°cia" in comando:
        falar("Abrindo a farm√°cia mais pr√≥xima no Google Maps...")
        webbrowser.open("https://www.google.com/maps/search/farm√°cia+perto+de+mim/")

    elif "sair" in comando:
        falar("At√© logo!")
        return False

    else:
        falar("N√£o entendi o comando.")

    return True
```

```python
falar("Ol√°, eu sou sua assistente virtual. Como posso ajudar?")

ativo = True
while ativo:
    comando = ouvir()
    if comando:
        ativo = executar_comando(comando)
```
